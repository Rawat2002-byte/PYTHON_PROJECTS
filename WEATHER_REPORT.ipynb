```python
top = data["Weather"].value_counts().head(10)
print(top)
```

```python
ab = sns.countplot(x="Weather",data =data,order=top.index)
sns.set(rc={"figure.figsize":(20,5)})

for bars in ab.containers:
    ab.bar_label(bars)
```

*FROM THE GRAPH THE MAINLY CLEAR IS 2106 TIMES , MOSTLY CLOUDLY IS 2069 TIMES , CLOUDY IS 1728 TIMES, CLEAR 1326 TIMES , SNOW IS 390 TIMES*
